```python
# Read in the files
google_data = pd.read_csv('Resources/googleplaystore.csv')
apple_data = pd.read_csv('Resources/AppleStore.csv')

google_data.drop_duplicates(subset='App', inplace=True)
google_data = google_data[google_data['Android Ver'] != np.nan]
google_data = google_data[google_data['Android Ver'] != 'NaN']
google_data = google_data[google_data['Installs'] != 'Free']
google_data = google_data[google_data['Installs'] != 'Paid']

google_data.head(20)
```

##Data Cleaning

Convert all app sizes to MB Remove '+' from 'Number of Installs' to make it numeric Convert all review text to English language using Google Translator library

```python
# - Installs : Remove + and ,

google_data['Installs'] = google_data['Installs'].apply(lambda x: x.replace('+', '') if '+' in str(x) else x)
google_data['Installs'] = google_data['Installs'].apply(lambda x: x.replace(',', '') if ',' in str(x) else x)
google_data['Installs'] = google_data['Installs'].apply(lambda x: int(x))
```

```python
# Size : Replace 'Varies with device to NaN value Remove 'M', Replace 'k' and divide by 10^-3


google_data['Size'] = google_data['Size'].apply(lambda x: str(x).replace('Varies with device', 'NaN') if 'Varies with device' in str(x) else x)

# Remove 'M' and ',' and 'K'
google_data['Size'] = google_data['Size'].apply(lambda x: str(x).replace('M', '') if 'M' in str(x) else x)
google_data['Size'] = google_data['Size'].apply(lambda x: str(x).replace(',', '') if 'M' in str(x) else x)
google_data['Size'] = google_data['Size'].apply(lambda x: float(str(x).replace('k', '')) / 1000 if 'k' in str(x) else x)
google_data['Size'] = google_data['Size'].apply(lambda x: float(x))

google_data['Installs'] = google_data['Installs'].apply(lambda x: float(x))

# Remove '$'
google_data['Price'] = google_data['Price'].apply(lambda x: str(x).replace('$', '') if '$' in str(x) else str(x))
google_data['Price'] = google_data['Price'].apply(lambda x: float(x))

google_data['Reviews'] = google_data['Reviews'].apply(lambda x: int(x))
```

## look for any evident patterns or relationships between the features.

```python
x = google_data['Rating'].dropna()
z = google_data['Installs'][google_data.Installs!=0].dropna()
p = google_data['Reviews'][google_data.Reviews!=0].dropna()
t = google_data['Type'].dropna()
price = google_data['Price']

p = sns.pairplot(pd.DataFrame(list(zip(x, np.log(z), np.log10(p), t, price)), 
                        columns=['Rating', 'Installs', 'Reviews', 'Type', 'Price']), hue='Type', palette="Set2")
```

We can break this down even further to inspect if we have categories which perform exceptionally good or bad.

## App ratings across categories - One Way Anova Test

```python
import scipy.stats as stats
graphs = stats.graphs_oneway(google_data.loc[google_data.Category == 'BUSINESS']['Rating'].dropna(), 
               google_data.loc[google_data.Category == 'FAMILY']['Rating'].dropna(),
               google_data.loc[google_data.Category == 'GAME']['Rating'].dropna(),
               google_data.loc[google_data.Category == 'PERSONALIZATION']['Rating'].dropna(),
               google_data.loc[google_data.Category == 'LIFESTYLE']['Rating'].dropna(),
               google_data.loc[google_data.Category == 'FINANCE']['Rating'].dropna(),
               google_data.loc[google_data.Category == 'EDUCATION']['Rating'].dropna(),
               google_data.loc[google_data.Category == 'MEDICAL']['Rating'].dropna(),
               google_data.loc[google_data.Category == 'TOOLS']['Rating'].dropna(),
               google_data.loc[google_data.Category == 'PRODUCTIVITY']['Rating'].dropna()
              )

print(graphs)
print('\nThe p-value is extremely small, hence we reject the null hypothesis in favor of the alternate hypothesis.\n')
#temp = df.loc[df.Category.isin(['BUSINESS', 'DATING'])]

groups = google_data.groupby('Category').filter(lambda x: len(x) > 286).reset_index()
array = groups['Rating'].hist(by=groups['Category'], sharex=True, figsize=(20,20))
```
