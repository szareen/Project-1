```python
# Read in the files
google_data = pd.read_csv('Resources/googleplaystore.csv')
apple_data = pd.read_csv('Resources/AppleStore.csv')
```

```python
#Re-organizing dataframe to only display specific columns

google_df = google_data[["App", "Category", "Rating", "Reviews", "Size", "Installs", "Type", "Price", "Genres"]]
apple_df = apple_data[["track_name","size_bytes","price","rating_count_tot","user_rating","prime_genre"]]
```

```python
google_df.drop_duplicates(subset='App', inplace=True)
google_df = google_df[google_df['Installs'] != 'Free']
google_df = google_df[google_df['Installs'] != 'Paid']

google_df.head(20)
```

Data Cleaning

Convert all app sizes to MB Remove '+' from 'Number of Installs' to make it numeric Convert all review text to English language using Google Translator library

```python
# - Installs : Remove + and ,

google_df['Installs'] = google_df['Installs'].apply(lambda x: x.replace('+', '') if '+' in str(x) else x)
google_df['Installs'] = google_df['Installs'].apply(lambda x: x.replace(',', '') if ',' in str(x) else x)
google_df['Installs'] = google_df['Installs'].apply(lambda x: int(x))
```

```python
# Size : Replace 'Varies with device to NaN value Remove 'M', Replace 'k' and divide by 10^-3


google_df['Size'] = google_df['Size'].apply(lambda x: str(x).replace('Varies with device', 'NaN') if 'Varies with device' in str(x) else x)

# Remove 'M' and ',' and 'K'
google_df['Size'] = google_df['Size'].apply(lambda x: str(x).replace('M', '') if 'M' in str(x) else x)
google_df['Size'] = google_df['Size'].apply(lambda x: str(x).replace(',', '') if 'M' in str(x) else x)
google_df['Size'] = google_df['Size'].apply(lambda x: float(str(x).replace('k', '')) / 1000 if 'k' in str(x) else x)
google_df['Size'] = google_df['Size'].apply(lambda x: float(x))

google_df['Installs'] = google_df['Installs'].apply(lambda x: float(x))

# Remove '$'
google_df['Price'] = google_df['Price'].apply(lambda x: str(x).replace('$', '') if '$' in str(x) else str(x))
google_df['Price'] = google_df['Price'].apply(lambda x: float(x))

google_df['Reviews'] = google_df['Reviews'].apply(lambda x: int(x))
```

```python
#Renaming the columns and adding "Type" column in order to categorize the free and paid apps

apple_df = apple_df.rename(columns = {"track_name":"App","rating_count_tot":"Rating Count",
                                          "user_rating":"Rating","prime_genre":"Genres","price":"Price"})
apple_df["Size (M)"] = (apple_df["size_bytes"]*(1e-6)).map("{:.1f}".format)
apple_df = apple_df.drop(columns = "size_bytes")
apple_df["Type"]=""
apple_df.loc[apple_df["Price"]==0,["Type"]]="Free"
apple_df.loc[apple_df["Type"]=="",["Type"]]="Paid"
apple_df.head()
```

```python
google_df["Price"] = pd.to_numeric(google_df["Price"]).map("{:.1f}".format)
google_df = google_df.rename(columns={"Size":"Size (M)"})
google_df["Category"] = google_df["Category"].str.title()
google_df["Category"] = google_df["Category"].replace({"_And_":" & ","_":" "},regex=True)
google_df = google_df.reset_index(drop=True)
google_df.head()
```

```python
google_df["Category"].unique()
```

```python
google_df["New Genres"]=""
google_df.loc[google_df["Category"]=="Game",["New Genres"]]="Games"

google_df.loc[(google_df["Category"]=="Education")|
           (google_df["Category"]=="Books & Reference")|
           (google_df["Category"]=="Libraries & Demo"),["New Genres"]]="Education"

google_df.loc[(google_df["Category"]=="Entertainment")|
           (google_df["Category"]=="Events")|
           (google_df["Category"]=="Social")|
           (google_df["Category"]=="Sports")|
           (google_df["Category"]=="News & Magazines"),["New Genres"]]="Entertainment"

google_df.loc[(google_df["Category"]=="Video Players")|
           (google_df["Category"]=="Photography"),["New Genres"]]="Photo & Video"

google_df.loc[(google_df["Category"]=="Tools")|
           (google_df["Category"]=="Communication")|
           (google_df["Category"]=="Personalization")|
           (google_df["Category"]=="Weather")|
           (google_df["Category"]=="Maps & Navigation")|
           (google_df["Category"]=="Finance"),["New Genres"]]="Utilities"

google_df.loc[(google_df["Category"]=="Productivity")|
           (google_df["Category"]=="Art & Design")|
           (google_df["Category"]=="Business"),["New Genres"]]="Productivity"

google_df.loc[(google_df["Category"]=="Health & Fitness")|
           (google_df["Category"]=="Medical"),["New Genres"]]="Health & Fitness"

google_df["New Genres"].unique()
```

```python
categories = ["Education", "Entertainment", "Health & Fitness", "Games", "Productivity", "Photo & Video", "Utilities"]
games = google_df.loc[google_df["New Genres"]=="Games"]["Genres"]

family = google_df.loc[google_df["Category"]=="Family"]["Genres"].str.split(";", expand=True)[0].to_frame()

family[0] = family[0].replace({"Educational":"Education"})

for index, row in family.iterrows():
    if row.isin(categories).any():
        google_df.iloc[index,8]=row[0]
    elif row.isin(games).any():
        google_df.iloc[index,8]="Games"
    elif row[0] == "Art & Design":
        google_df.iloc[index,8]="Productivity"
    elif row[0] == "Video Players & Editors":
        google_df.iloc[index,8]="Photo & Video"
    elif row[0] == "Communication":
        google_df.iloc[index,8]="Utilities"
    elif row[0] == "Books & Reference":
        google_df.iloc[index,8]="Education"

google_df.loc[google_df["New Genres"]=="",["New Genres"]]="Other"

google_cat = google_df.groupby("New Genres").count().sort_values("App",ascending=False)

google_cat
```

```python
apple_df["New Genres"]=""
apple_df.loc[apple_df["Genres"]=="Games",["New Genres"]]="Games"

apple_df.loc[(apple_df["Genres"]=="Education")|
          (apple_df["Genres"]=="Book")|
          (apple_df["Genres"]=="Reference"),["New Genres"]]="Education"

apple_df.loc[(apple_df["Genres"]=="Entertainment")|
          (apple_df["Genres"]=="Social Networking")|
          (apple_df["Genres"]=="Sports")|
          (apple_df["Genres"]=="News"),["New Genres"]]="Entertainment"

apple_df.loc[apple_df["Genres"]=="Photo & Video",["New Genres"]]="Photo & Video"
apple_df.loc[(apple_df["Genres"]=="Utilities")|
          (apple_df["Genres"]=="Weather")|
          (apple_df["Genres"]=="Navigation")|
          (apple_df["Genres"]=="Finance"),["New Genres"]]="Utilities"

apple_df.loc[(apple_df["Genres"]=="Productivity")|
          (apple_df["Genres"]=="Business"),["New Genres"]]="Productivity"

apple_df.loc[(apple_df["Genres"]=="Health & Fitness")|
          (apple_df["Genres"]=="Medical"),["New Genres"]]="Health & Fitness"

apple_df.loc[apple_df["New Genres"]=="",["New Genres"]]="Other"

apple_cat = apple_df.groupby("New Genres").count().sort_values("App",ascending=False)
apple_cat
```
